```python
def power_iteration(A, x_0, p, num_iterations):
    
    import numpy as np
    
    
    # A: nxn matrix, x_0: initial guess, p: type of norm
    x_0 = x_0/np.linalg.norm(x_0,p)
    x_k = x_0
    for i in range(num_iterations):
        y_k = A @ x_k
        x_k = y_k/np.linalg.norm(y_k,p)
    return x_k

def inverse_power_iteration(A, x_0, p, num_iterations,shift=0):

    import scipy.linalg as sla
    import numpy as np
    import numpy.linalg as la
    
    A = A - shift*np.identity(A.shape[0])
    x = x_0/la.norm(x_0,p)
    P,L,U = sla.lu(A)
    for k in range(num_iterations):
        y = sla.solve_triangular(L,np.dot(P.T,x),lower=True)
        x = sla.solve_triangular(U,y)
        x = x/la.norm(x,p)
    return x
```

```python
def central_finite(f,x,h):
    
    import numpy as np
    
    fx = f(x)
    df = np.zeros(len(x))
    
    for i in range(len(x)):
        xcopy1 = x.copy()
        xcopy2 = x.copy()
        xcopy1[i] += h
        xcopy2[i] -= h
        fxph = f(xcopy1)
        fxmh = f(xcopy2)
        df[i] = (fxph-fxmh) / (2*h)
        
    return df

def forward_finite(f,x,h):
    
    import numpy as np
    
    fx = f(x)
    df = np.zeros(len(x))
    
    for i in range(len(x)):
        
        xcopy = x.copy()
        xcopy[i] += h
        fxph = f(xcopy)
        df[i] = (fxph - fx) / h
    
    return df

def backward_finite(f,x,h):
    
    import numpy as np
    
    fx = f(x)
    df = np.zeros(len(x))
    
    for i in range(len(x)):
        
        xcopy = x.copy()
        xcopy[i] -= h
        fxmh = f(xcopy)
        df[i] = (fx - fxmh) / h
    
    return df
```

```python
def golden_search(f,a,b,tol):
    import numpy as np
    brackets = []
    h = b-a
    t = (np.sqrt(5)-1)/2 #.618
    x1 = a+(1-t)*h
    x2 = a + t*h
    f1 = f(x1)
    f2 = f(x2)

    while h >= tol:

        if f1<f2:

            b = x2
            x2 = x1
            f2 = f1
            h *= t
            x1 = a + (1-t)*h
            f1 = f(x1)

        if f1 > f2:

            a = x1
            x1 = x2
            f1=f2
            h*=t
            x2=a+t*h
            f2 = f(x2)

        brackets.append([a, x1, x2, b])
        
    return brackets
```

```python
## Finds the root of a function. Simplest method. O(1)

def bisection(a,b,f,n_iter,epsilon):
    
    import numpy as np
    roots = []
    m = (a+b)/2

    for j in range(n_iter):
        
        try:
            assert a < b
            assert f(a)*f(b) < 0
            
        except:
            break
            
        if np.abs(f(m)) < epsilon:
            roots.append(m)
            break
        
        if f(a)*f(m) < 0:
            b = m
            
        elif f(b)*f(m) < 0:
            a = m
            
        m = (a+b)/2
        
        roots.append([a,m,b])
        
    return roots
```
